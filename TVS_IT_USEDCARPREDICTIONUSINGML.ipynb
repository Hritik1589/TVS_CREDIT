1. Context
  What determines the price of a used car?

  In fact, the car's value drops by his 20% of its original value in his first year.

  The make and model of the car, total mileage, overall condition of the vehicle, and various other factors further affect the resale value of the car.

2. Problem
New car prices in the industry are set by the manufacturers and incur some additional costs to the government in the form of taxes. In this way, a customer buying a new car can be confident that the money invested is reasonable. However, used car sales are increasing worldwide due to rising new car prices and the inability of customers to buy new cars due to lack of funds (Pal, Arora, and Palakurthy, 2018). You need a used car price prediction system to effectively determine a car's value using various attributes. Some websites offer this service, but their forecasting methods may not be optimal. Various models and systems also help predict the actual market value performance of used vehicles. When buying or selling, it is important to know the actual market value.

3. Data Description
This dataset contains data on over 7000 real value vehicles ready to take on another owner in all major Tier 1 and Tier 2 cities of India . The information includes the car make, model, fuel type, year of manufacture, etc.

### **8.Handling missing values**

The data body_type,transmission,source,car_availability,car_rating,ad_created_on,fitness_certificate,registered_city,registered_state contains null values. The next step is to handle missing data

```python
df1.dtypes
```

```python
for i in df.columns:
  print(i)
  print(df[i].unique())
  print("*********************************************************************************************")
```

Here we replace missing values ​​in object type columns with mode and numeric type columns with mean

```python
category_columns=df1.select_dtypes(include=['object']).columns.tolist()
integer_columns=df1.select_dtypes(include=['int64','float64']).columns.tolist()

for column in df1:
    if df1[column].isnull().any():
        if(column in category_columns):
            df1[column]=df1[column].fillna(df1[column].mode()[0])
        else:
            df1[column]=df1[column].fillna(df1[column].mean)
```

## **14.Model Evaluation**

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error
def metric(y_test,y_predict):
    mae=mean_absolute_error(y_test,y_predict) #mean_absolute_error
    mse=mean_squared_error(y_test,y_predict) #mean_squared_error
    rmse=mean_squared_error(y_test,y_predict,squared=False)
    return [mae,mse,rmse]
```
